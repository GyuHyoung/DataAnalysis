# 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
tips = sb.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips['total_bill'], bins=10) #bins는 x값의 간격
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

```python
#이변량(산점도) 그래프
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
#박스 그래프
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot(

    (tips[tips['sex'] == 'Female']['tip'],tips[tips['sex'] == 'Male']['tip']),
    labels = ['Female','Male'])
    
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

```python
def recode_sex(sex):
    if sex == 'Female' :
        return 0
    else :
        return 1
```

```python
#sex열에 recode_sex 함수 브로드캐스팅
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10, #점의 크기로 테이블당 인원수 나타냄
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tips Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

# 단변량 그래프 그리기 - 히스토그램

```python
import seaborn as sb

tips = sb.load_dataset("tips")
```

```python
ax = plt.subplots()
ax = sb.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplots()
ax = sb.distplot(tips['total_bill'], kde=False) #밀집도 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sb.distplot(tips['total_bill'], hist=False) #밀집도만 표시
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
hist_den_rug, ax = plt.subplots()
ax = sb.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
ax = plt.subplots()
ax = sb.countplot('day',data=tips) #이산값 표시
ax.set_title('Count of days')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sb.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```
